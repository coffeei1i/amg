```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

row_data_path = "all_row_InN.csv"
row_data = pd.read_csv(row_data_path)
row_data .rename(columns={row_data .columns[0]: "column"}, inplace=True)
df = row_data 
desired_order = [
    "LA","BA","BMAp","COAp","PA","BMAa","COAa",
    "MEAav","MEApv","MEAad","MEApd","CEAm","CEAl","IA"
]


desired_order = [c for c in desired_order if c in row_data .columns]


new_cols = ["column"] + desired_order


row_data = row_data[new_cols]
order = [
    "InN_Isl1_Kcnh8",
    "InN_Nxph2_Foxp2",
    "InN_Abcb5_Pax6",
    "InN_Calcr_Esr2",
    "InN_Esr2_Cyp19a1",
    "InN_Lhx8_Th",
    "InN_Gal_Dock8",
    "InN_Tshr_Qrfprl",
    "InN_Apob_Prkcq",
    "InN_Sst_Ntn1",
    "InN_Sst_Myzap",
    "InN_Sst_Edaradd",
    "InN_Pvalb_Adamts5",
    "InN_Pvalb_Vipr2",
    "InN_Lamp5_Ndnf",
    "InN_Lamp5_Id2",
    "InN_Ndnf_Ppp1r1c",
    "InN_Htr3a_Glyctk",
    "InN_Htr3a_Frem1",
    "InN_Vip_Crh",
    "InN_Prkcd_Chodl",
    "InN_Drd2_Adora2a",
    "InN_Sst_Vipr2",
    "InN_Tshz1_Lamp5",
    "InN_Tshz1_Drd1"
]


row_data = row_data.set_index("column")

order_present = [x for x in order if x in row_data.index]
row_data = row_data.loc[order_present]
row_data.reset_index(inplace=True)
```

```python
area_names = np.array([
    "LA","BA","BMAp","COAp","PA","BMAa","COAa",
    "MEAav","MEApv","MEAad","MEApd","CEAm","CEAl","IA"
])
```

```python
# Extracting percentage data and cell type labels
percentages = row_data.iloc[0:, 1:].astype(float).values  # converting string percentages to float
cell_types = row_data.iloc[0:, 0].values


# Importing matplotlib library for visualization
import matplotlib.pyplot as plt
import random

def random_hex():
    return '{:06x}'.format(random.randint(0, 0xFFFFFF))

area_color_map = {area: random_hex() for area in area_names}
area_color_map = {
    'IA':    '64b5f6',
    'CEAl':  '036eb8',
    'CEAm':  '02d8d0',
    'MEApd': '006934',
    'MEAad': '00913A',
    'MEApv': '8FC31F',
    'MEAav': '13AE67',
    'COAa':  'FFE200',
    'BMAa':  'F39800',
    'PA':    'E60012',
    'COAp':  'FFF100',
    'BMAp':  'E83828',
    'BA':    'EB6100',
    'LA':    'F8B62D',
}

# Creating the stacked bar chart
fig, ax = plt.subplots(figsize=(12, 8))
bottom = np.zeros(len(cell_types))

for area, color in area_color_map.items():
    area_data = percentages[:, list(row_data.columns.values[1:]).index(area)]
    ax.bar(cell_types, area_data, label=area, bottom=bottom, color=f'#{color}')
    bottom += area_data

ax.set_xlabel('Cell Type')
ax.set_ylabel('Percentage')
ax.set_title('Distribution of Cell Types Across Different Areas')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], title='Area', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()


plt.show()
```
